# Regularization with SciKit-Learn

Regularization attempts to minimize the RSS (residual sum of squares) *and* a penalty factor. This penalty factor will penalize models that have coefficients that are too large. Some methods of regularization will actually cause non useful features to have a coefficient of zero, in which case the model does not consider the feature.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from google.colab import drive
drive.mount('/content/gdrive')
```

```python
df = pd.read_csv('gdrive/My Drive/csv_files/Advertising.csv')
```

```python
df.head()
```

```python
X = df.drop('sales',axis=1)
```

```python
y = df['sales']
```

### Polynomial Conversion

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
polynomial_converter = PolynomialFeatures(degree=3,include_bias=False)
```

```python
poly_features = polynomial_converter.fit_transform(X)
```

```python
X.shape
```

```python
poly_features.shape
```

### Train | Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(poly_features, y, test_size=0.3, random_state=101)
```

```python
X_train.shape
```

----
----

## Scaling the Data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
X_train[0]
```

```python
poly_features[0]
```

## Ridge Regression

```python
from sklearn.linear_model import Ridge
```

```python
#help(Ridge)
```

```python
ridge_model = Ridge(alpha=10)
```

```python
ridge_model.fit(X_train,y_train)
```

```python
test_predictions = ridge_model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
MAE = mean_absolute_error(y_test,test_predictions)
```

```python
MAE
```

```python
RMSE = np.sqrt(mean_squared_error(y_test,test_predictions))
```

```python
RMSE
```

### Choosing an alpha value with Cross-Validation


```python
from sklearn.linear_model import RidgeCV       # Ridge Cross Validation 
```

```python
ridge_cv_model = RidgeCV(alphas= (0.1, 1, 10))
```

```python
ridge_cv_model.fit(X_train,y_train)
```

```python
ridge_cv_model.alpha_
```

```python
from sklearn.metrics import SCORERS
```

```python
SCORERS.keys()
```

```python
ridge_cv_model = RidgeCV(alphas= (0.1, 1, 10),scoring='neg_mean_squared_error')
```

```python
ridge_cv_model.fit(X_train,y_train)
```

```python
ridge_cv_model.alpha_
```

```python
test_predictions = ridge_cv_model.predict(X_test)
```

```python
MAE = mean_absolute_error(y_test,test_predictions)
```

```python
RMSE = np.sqrt(mean_squared_error(y_test,test_predictions))
```

```python
MAE
```

```python
RMSE
```

```python
ridge_cv_model.coef_
```

```python
ridge_cv_model.best_score_
```

## LASSO Regression

LASSO - Least Absolute Shrinkage and Selection Operator

```python
from sklearn.linear_model import LassoCV
```

```python
lasso_cv_model = LassoCV(eps=0.1, n_alphas=100,cv=5)
```

```python
lasso_cv_model.fit(X_train,y_train)
```

```python
lasso_cv_model.alpha_
```

```python
test_predictions = lasso_cv_model.predict(X_test)
```

```python
MAE = mean_absolute_error(y_test,test_predictions)
```

```python
RMSE = np.sqrt(mean_squared_error(y_test,test_predictions))
```

```python
MAE
```

```python
RMSE
```

```python
lasso_cv_model.coef_
```

```python
lasso_cv_model = LassoCV(eps=0.001, n_alphas=100,cv=5,max_iter=1000000)
```

```python
lasso_cv_model.fit(X_train,y_train)
```

```python
lasso_cv_model.alpha_
```

```python
test_predictions = lasso_cv_model.predict(X_test)
```

```python
MAE = mean_absolute_error(y_test,test_predictions)
```

```python
RMSE = np.sqrt(mean_squared_error(y_test,test_predictions))
```

```python
MAE
```

## Elastic Net

Elastic Net combines the penalties of ridge regression and lasso in an attempt to get the best of both worlds!

```python
from sklearn.linear_model import ElasticNetCV
```

```python
elastic_model = ElasticNetCV(l1_ratio=[.1, .5, .7,.9, .95, .99, 1],eps=0.001,n_alphas=100,max_iter=1000000)
```

```python
elastic_model.fit(X_train,y_train)
```

```python
elastic_model.l1_ratio
```

```python
elastic_model.l1_ratio_
```

```python
elastic_model.alpha_
```

```python
lasso_cv_model.alpha_
```

```python
test_predictions = elastic_model.predict(X_test)
```

```python
MAE = mean_absolute_error(y_test,test_predictions)
```
